Actverse_python project examples and usage

Actverse_python project is used to extract meaningful data from csv files provided by AVATAR or other video tracking software.

There are currently 3 modules

FileManager module involves importing data from csv file to a the dataframe that can be processed with python
 - adv_load: allows users to use windows gui to select files or folders to import
 - csvload: actual importing of the CSV files
 - preprocess: extracting certain rows/columns or arbitrary center point.


Calculator.simple moudule involves simple calculations 
  - Postion differences from frame to frame
  - Angular differences
  - Fixation(linear transformation) of bodyparts
  - Whether subject is at the center of the apparatus 
  - Body length

Plotter module involves plotting 
  - avatarheatmap: shows the heatmap
  - hist: shows histogram

To start, you first need to import data from csv to pandas dataframe. 
You can do this easily with Easyavatarload function or manually with avatarcsvloader or generalcsvloader in csvload

```python
from FileManager import adv_load
from FileManager import csvload

testdata = csvload.avatarcsvloader('example.csv')
```

Afterward, you can use the custom functions in Calcuator module to calculate statistics you might need.

```python
from Calculator.Simple import Angular
from Calculator.Simple import L2norm

bl1 = L2norm.bodylength(testdata)
mv1 = L2norm.movement(testdata)
angle1 = Angular.point2lineangle(testdata.loc[:,['Nose_x','Nose_y','Nose_z','Head_x','Head_y','Head_z','Bodycenter_x','Bodycenter_y','Bodycenter_z']])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(0)
sns.lineplot(data=angle1)
plt.figure(1)
sns.lineplot(data=mv1)
plt.figure(2)
sns.lineplot(data=bl1)
```

Plotter contains basic plots you can use with the analysis. 

```python
from Plotter import Basic
Basic.avatarheatmap(testdata.loc[:,['Bodycenter_x']],nofgrid=6)
```
